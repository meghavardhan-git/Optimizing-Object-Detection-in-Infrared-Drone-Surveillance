```python
from openvino.runtime import Core
import cv2
import numpy as np
import time

# Initialize OpenVINO runtime
ie = Core()

# Load your converted YOLOv8 model
model_path = "/content/openvino_model/best.xml"  # ✅ Updated path
model = ie.read_model(model_path)
compiled_model = ie.compile_model(model, device_name="CPU")

# Get input and output layer names
input_layer = compiled_model.input(0)
output_layer = compiled_model.output(0)

# Load and preprocess image
image_path = "/content/hit-uav/hit-uav/images/test/0_120_60_0_09554.jpg"  # ✅ Updated image path for Colab
image = cv2.imread(image_path)
input_image = cv2.resize(image, (640, 640))  # Resize to model input size
input_image = input_image.transpose(2, 0, 1)  # HWC to CHW
input_image = np.expand_dims(input_image, axis=0)  # Add batch dimension
input_image = input_image.astype(np.float32) / 255.0  # Normalize to [0, 1]

# Run inference
start = time.time()
outputs = compiled_model([input_image])[output_layer]
end = time.time()

print(f"✅ OpenVINO Inference time: {end - start:.4f} seconds")

# Output shape
print("Output shape:", outputs.shape)
```

```python
# 📌 YOLOv8 Evaluation & Metrics Collection Script
import time, os, cv2, numpy as np, torch
import pandas as pd
from ultralytics import YOLO
from openvino.runtime import Core

# ✅ Load trained YOLOv8 model
model = YOLO("/content/runs/detect/yolov8_hit-uav-new/weights/best.pt")

# -------------------------
# 1. Validation Metrics
# -------------------------
results = model.val()
metrics = {
    "mAP@0.5:0.95": results.box.map,
    "mAP@0.5": results.box.map50,
    "mAP@0.75": results.box.map75,
    "Precision": results.box.mp,
    "Recall": results.box.mr,
}

# Per-class AP
for i, class_map in enumerate(results.box.maps):
    metrics[f"AP_{model.names[i]}"] = class_map

# -------------------------
# 2. Efficiency (PyTorch)
# -------------------------
# Find a sample image in the test set
test_image_dir = "/content/hit-uav/hit-uav/images/test"
test_images = [os.path.join(test_image_dir, f) for f in os.listdir(test_image_dir) if f.endswith('.jpg')]

if test_images:
    img_path = test_images[0] # Use the first image found
    start = time.time()
    _ = model(img_path)
    end = time.time()
    metrics["PyTorch Inference Time (s)"] = end - start
    metrics["PyTorch FPS"] = 1.0 / (end - start)
else:
    print("No test images found in the specified directory for PyTorch inference.")
    metrics["PyTorch Inference Time (s)"] = None
    metrics["PyTorch FPS"] = None


# -------------------------
# 3. Export to ONNX + OpenVINO
# -------------------------
onnx_path = model.export(format="onnx")  # Export ONNX
!python -m openvino.tools.mo --input_model /content/runs/detect/yolov8_hit-uav-new/weights/best.onnx --output_dir openvino_model

ie = Core()
ov_model = ie.read_model("/content/openvino_model/best.xml") # Use best.xml
compiled_model = ie.compile_model(ov_model, device_name="CPU")

input_layer = compiled_model.input(0)
output_layer = compiled_model.output(0)

# Preprocess image for OpenVINO
if test_images:
    image = cv2.imread(img_path)
    input_image = cv2.resize(image, (640, 640))
    input_image = input_image.transpose(2, 0, 1)[None].astype(np.float32) / 255.0

    # Inference with OpenVINO
    start = time.time()
    _ = compiled_model([input_image])[output_layer]
    end = time.time()

    metrics["OpenVINO Inference Time (s)"] = end - start
    metrics["OpenVINO FPS"] = 1.0 / (end - start)
else:
     print("No test images found for OpenVINO inference.")
     metrics["OpenVINO Inference Time (s)"] = None
     metrics["OpenVINO FPS"] = None


# -------------------------
# 4. Model Size & Params
# -------------------------
pt_size = os.path.getsize("/content/runs/detect/yolov8_hit-uav-new/weights/best.pt") / 1e6
onnx_size = os.path.getsize("/content/runs/detect/yolov8_hit-uav-new/weights/best.onnx") / 1e6 # Use best.onnx
xml_size = os.path.getsize("/content/openvino_model/best.xml") / 1e6 # Use best.xml

metrics["YOLOv8 PT Size (MB)"] = pt_size
metrics["YOLOv8 ONNX Size (MB)"] = onnx_size
metrics["YOLOv8 OpenVINO Size (MB)"] = xml_size
metrics["YOLOv8 Parameters (M)"] = sum(p.numel() for p in model.model.parameters()) / 1e6

# -------------------------
# 5. Save Results
# -------------------------
df = pd.DataFrame([metrics])
df.to_csv("yolov8_results.csv", index=False)

print("✅ YOLOv8 evaluation complete. Results saved to yolov8_results.csv")
print(df)
```
